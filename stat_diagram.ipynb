```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

```

```python
df = pd.read_csv('mental-heath-in-tech-2016_20161114.csv')
df_stat = df.copy()
# Recode Comp size & country columns (for ease when doing plots)
df_stat['How many employees does your company or organization have?'].replace(to_replace = ['More than 1000'], value = '>1000', inplace = True)
df_stat['What country do you work in?'].replace(to_replace = ['United States of America'], value = 'USA', inplace = True)
df_stat['What country do you work in?'].replace(to_replace = ['United Kingdom'], value = 'UK', inplace = True)
```

```python
df_stat = df_stat[df_stat['What country do you work in?'].isin(['USA','UK','Canada', 'Germany', 'Netherlands','Australia'])]
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
plt.figure(figsize = (16,4))
fig.set_figheight(5)
fig.set_figwidth(20)
plt.subplots_adjust(wspace = 0)
# plt.suptitle('Main Title')
# No of respondents by Country
sns.countplot(x = df_stat['What country do you work in?'], hue = df_stat['Have you ever sought treatment for a mental health issue from a mental health professional?'], ax=ax1)
ax1.set_title('No. of Respondents by Country and sought treatment', fontsize = 16)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=15, ha="right")
ax1.set_xlabel('Country', fontsize = 18)
ax1.set_ylabel('Count', fontsize = 18)
ax1.legend(['Not sought treatment', 'sought treatment'])
# No of respondents by Company Size
sns.countplot(x = df_stat['How many employees does your company or organization have?'], hue = df_stat['Have you ever sought treatment for a mental health issue from a mental health professional?'], ax=ax2, 
              order = ['1-5', '6-25', '26-100', '100-500', '500-1000', '>1000'])
ax2.set_title('No. of Respondents by Company and sought treatment', fontsize = 16)
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=15, ha="right")
ax2.set_xlabel('Company Size', fontsize = 18)
ax2.set_ylabel('Count', fontsize = 18)
ax2.legend(['Not sought treatment', 'sought treatment']);
fig.savefig('C:/Users/afshi/OneDrive/Documents/unsuper/final/final/output.png')
```
